# Data Cleaning:

## 1. Removing the Twitter Handles(@user)

```python
# This is a gerneralized function which will look for a specific pattern in the given string and replace that pattern with
#  a blank space:

import re
def patter_finder(input_text, pattern):
    r = re.findall(pattern, input_text)
    for i in r:
        input_text = re.sub(i, '', input_text)
    
    return input_text
    
    
```

```python
tweets['tidy_tweet'] = np.vectorize(patter_finder)(tweets['text'],'@[\w]*')
```

```python
tweets.head(10)
```

## 2. Replacing the punctuations,numbers and special characters:

```python
tweets['tidy_tweet'] = tweets['tidy_tweet'].str.replace('[^a-zA-Z#]'," ")
```

## 3. Removing the short words(less than  4 character):

```python
#Here we remove words like 'are','is' etc which doesn't add much value to the tweet other than helping in
# creating a grammtically correct sentence:

tweets['tidy_tweet'] = tweets['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

## 4. Tokenization of the strings:

```python
## Tokenization is the process of splitting the tweets into individual words.

tokenized_tweet = tweets['tidy_tweet'].apply(lambda x : x.split())
```

```python
tokenized_tweet.head()
```

## 5. Stemming:

```python
##Stemming of the tokens:reducing the words to their root form( for eg:root form of playing is play)

import nltk
from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x:[stemmer.stem(i) for i in x])
```

```python
## Joining the tokens into complete tweets:

for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

tweets['tidy_tweet'] = tokenized_tweet
```

# Checking the presence of  Hashtags other than '#Article 370A'

```python
# This function will extract all the hastags from the tweets:

def Hastag_retriever(input_text):
    hastags = []
    for i in input_text:
        ht  = re.findall('#(\w+)', i)
        hastags.append(ht)
    return hastags
```

## 1. Hashtags used along with 'Article 370 A' for Neutral Tweets

```python
# Getting the Hastags for Neutral Tweets:
Regular_Tweets_hastags = Hastag_retriever(tweets[tweets['Sentiments'] == 'Neutral']['tidy_tweet'])
```

```python
# Since this will be a nested list as one single tweet can have mutiple hastags, so we have to unnest the list.
Regular_Tweets_hastags = sum(Regular_Tweets_hastags,[])
```

```python
a_1 = nltk.FreqDist(Regular_Tweets_hastags) 
# This caculates the Frequency of each words in the list.Since we have scrapped the data based on the Hastag 'Article 370A'
# There will be the presence of large number of Hastags 'Article 370A'
# We will exlude the Hastag 'Article 370A' while plotting the bar chart as we need to find the presence of other hashtags
# which were used along with the 'Article 370A' Hashtag

d_1 = pd.DataFrame({'Hashtag': list(a_1.keys()),
                   'Count': list(a_1.values())})

# Removing the the frequency of article 370A from the dataframe
d_1.drop(index = 3, inplace = True)

#selecting top 10 most frequent hashtags     
d_1 = d_1.nlargest(columns="Count", n = 10) 
plt.figure(figsize=(16,5))
ax = sns.barplot(data=d_1, y= "Hashtag", x = "Count")
plt.title("Top 10 Hashtags used along with the '#Article 370A' for Neutral Tweets", size = 20)
plt.show()
```
